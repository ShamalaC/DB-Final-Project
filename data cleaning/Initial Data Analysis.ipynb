```python
# Convert 'Released' to datetime
df_dropped['Released'] = pd.to_datetime(df_dropped['Released'], errors='coerce')  # errors='coerce' handles invalid dates

# Extract the year from the 'Released' column
df_dropped['Release Year'] = df_dropped['Released'].dt.year

# Filter for years 2021 and 2020
filtered_df = df_dropped[df_dropped['Release Year'].isin([2021,2020])]

# Now filtered_df contains only the rows where the release year is 2021 or 2020
filtered_df
```

```python
filtered_df.info()
```

```python
# Convert 'Released' to datetime
df_dropped['Released'] = pd.to_datetime(df_dropped['Released'], errors='coerce')

# Define start and end year-month
start_year, start_month = 2020, 9  # Start from May 2020
end_year, end_month = 2021, 5  # End at August 2021

# Create a boolean mask for the range
mask = (
    (df_dropped['Released'].dt.year > start_year) | 
    ((df_dropped['Released'].dt.year == start_year) & (df_dropped['Released'].dt.month >= start_month))
) & (
    (df_dropped['Released'].dt.year < end_year) | 
    ((df_dropped['Released'].dt.year == end_year) & (df_dropped['Released'].dt.month <= end_month))
)

# Apply the mask
filtered_df = df_dropped[mask]
```

```python
filtered_df.info()
```

```python
filtered_df.to_csv('filtered_on_released_date.csv', index=False)
```

```python
filtered_df.drop(['Developer Website', 'Developer Email','Privacy Policy','Currency','Scraped Time'], axis=1, inplace=True)
```
